- Построить график произвольной квадратичной функции

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)
y = 4 * x**2 + 2 * x + 3

plt.plot(x, y , color='green')
plt.title('Квадратичная функция: y = 4x^2 + 2x + 3')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True)
plt.show()
```

-  Построить график произвольной гармонической функции

```python
x_values = np.linspace(0, 15, 500)
amplitude = 0.5
frequency = 1
y_values = amplitude * np.sin(frequency * 2 * np.pi * x_values)

plt.plot(x_values, y_values, color='green', linestyle='-', label=f'{amplitude}sin({frequency} * 2πx)')
plt.title('Гармоническая функция:'' y=A⋅sin(2πfx)')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True)
plt.legend()
plt.show()
```

- Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

x = np.linspace(-10, 10, 400)
y = 4 * x**2 + 2 * x + 3
axs[0].plot(x, y, color='green')
axs[0].set_title('Квадратичная функция: y = 4x^2 + 2x + 3')
axs[0].set_xlabel('Ось X')
axs[0].set_ylabel('Ось Y')
axs[0].grid(True)

x_values = np.linspace(0, 15, 500)
amplitude = 0.5
frequency = 1
y_values = amplitude * np.sin(frequency * 2 * np.pi * x_values)
axs[1].plot(x_values, y_values, color='green', linestyle='-', label=f'{amplitude}sin({frequency} * 2πx)')
axs[1].set_title('Гармоническая функция: y=A⋅sin(2πfx)')
axs[1].set_xlabel('Ось X')
axs[1].set_ylabel('Ось Y')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

-  Построить графики п.2 и 3 в одной координатной плоскости

```python
x = np.linspace(-10, 10, 400)
y_quad = 4 * x**2 + 2 * x + 3

x_values = np.linspace(0, 15, 500)
amplitude = 0.5
frequency = 1
y_harmonic = amplitude * np.sin(frequency * 2 * np.pi * x_values)

plt.plot(x, y_quad, color='green', label='Квадратичная: y = 4x^2 + 2x + 3')
plt.plot(x_values, y_harmonic, color='blue', label=f'Гармоническая: y = {amplitude}sin({frequency} * 2πx)')

plt.title('Квадратичная и Гармоническая функции')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True)
plt.legend()

plt.show()
```

- Построить график функции, заданной параметрически

```python
t = np.linspace(0, 4 * np.pi, 400)
x = np.cos(3 * t)
y = np.sin(2 * t)

plt.plot(x, y)
plt.title('Параметрическая функция: x = cos(3t), y = sin(2t)')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True)
plt.show()
```

- Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму

```python
years = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]


plt.bar(years, values, color='skyblue')
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум, сом')
plt.xticks(rotation=45)  
plt.grid(axis='y')  
plt.tight_layout()  

plt.show()
```

- Построить круговой график 

```python

years = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]


plt.pie(values, labels=years, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')
plt.axis('equal')  

plt.show()
```

- Построить график произвольной каркасной поверхности

```python
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.cos(x) + np.sin(y)  #


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(x, y, z, color='green')


ax.set_title('Каркасная поверхность: z = cos(x) + sin(y)')
ax.set_xlabel('Ось X')
ax.set_ylabel('Ось Y')
ax.set_zlabel('Ось Z')

plt.show()
```
